```python
from statsmodels.stats.multitest import fdrcorrection
from scipy.stats import hypergeom

# hypergeometric test:
# We can then compute a probability of drawing X red marbles out of N from 
# a jar containing n red marbles out of M in the following way:
def hyper_test(x, M, n, N):
    pval = hypergeom.sf(x-1, M, n, N)
    #print(pval)
    return pval

def get_BH_correct_pval(df, pval_col):
    (reject_list, pval_bh) = fdrcorrection(df[pval_col], alpha=0.05, method='indep', is_sorted=False)
    return pval_bh

def add_unanno_col(df_enrich, all_col='ClusterMembers', in_col='ClusterMembersAnnotated', un_col='ClusterMembersUnannotated'):
    df = df_enrich.copy(deep=True)
    df[un_col] = np.nan
    for row in df.iterrows():
        all_genes = row[1][all_col].split(';')
        in_genes = row[1][in_col].split(';')
        un_genes = list(set(all_genes) - set(in_genes))
        un_genes = ';'.join(un_genes)
        df.loc[row[0], un_col] = un_genes
    return df
        
# GO enrichment function for clusters
# Update in May 2024: add back >=2 overlapping annotation requirement
def go_bp_enrichment_re2(df_clust, df_go_standard_matrix, fdr_cutoff=0.3, output_sub_dir='./', level='L2'):
    df_clust_new = df_clust.copy(deep=True)
    # Define universe as the genes used for clustering
    all_genes = ';'.join(df_clust_new['orf19'].tolist())
    all_genes = all_genes.split(';')
    num_all_genes = len(all_genes)
    print('Total genes covered by the clusters:', num_all_genes)
    all_genes_annotated = np.intersect1d(df_go_standard_matrix.index, all_genes)
    num_all_genes_annotated = len(all_genes_annotated)
    print('Total annotated genes covered by the clusters:', num_all_genes_annotated)
    df_go_standard_matrix_pop = df_go_standard_matrix.loc[all_genes_annotated]
    
    df_clust_new['num_enriched_GO'] = np.nan
    df_clust_new['enriched_GO'] = np.nan
    df_pval_list = []
    for row in df_clust_new.iterrows():
        orf19s = row[1]['orf19'].split(';')
        overlap_genes = np.intersect1d(df_go_standard_matrix_pop.index, orf19s)
        if len(overlap_genes) > 0:
            go_pval_list = []
            bg_list = []
            bg_sub_list = []
            sample_list = []
            sample_sub_list = []
            sample_sub_genes_list = []
            for go in df_go_standard_matrix_pop.columns:
                df_go = df_go_standard_matrix_pop[go]
                bg = len(df_go)
                bg_sub = len(df_go[df_go==1])
                df_go_sample = df_go_standard_matrix_pop.loc[overlap_genes, go]
                sample = len(df_go_sample)
                df_sample_sub = df_go_sample[df_go_sample==1]
                sample_sub = len(df_sample_sub)
                if sample_sub < 2:
                    go_pval_list.append(np.nan)
                    bg_list.append(np.nan)
                    bg_sub_list.append(np.nan)
                    sample_list.append(np.nan)
                    sample_sub_list.append(np.nan)
                    sample_sub_genes_list.append(np.nan)
                else:
                    sample_sub_genes = ';'.join(df_sample_sub.index.tolist())
                    pval = hyper_test(x=sample_sub, M=bg, n=bg_sub, N=sample)
                    go_pval_list.append(pval)
                    bg_list.append(bg)
                    bg_sub_list.append(bg_sub)
                    sample_list.append(sample)
                    sample_sub_list.append(sample_sub)
                    sample_sub_genes_list.append(sample_sub_genes)
                
            df_pval = pd.DataFrame(data={'GO':df_go_standard_matrix.columns, 
                                         '#Background':bg_list,
                                         '#Background_annotated':bg_sub_list,
                                         '#Cluster':sample_list,
                                         '#Cluster_annotated':sample_sub_list,
                                         'pval':go_pval_list, 'ClusterID':[str(row[0])]*len(go_pval_list),
                                        'ClusterMembers':[row[1]['orf19']]*len(go_pval_list),
                                        #'ClusterMembersInStandard':[';'.join(overlap_genes.tolist())]*len(go_pval_list)
                                        'ClusterMembersAnnotated':sample_sub_genes_list})
            #print(df_pval.shape)
            df_pval = df_pval[df_pval['ClusterMembersAnnotated'].notnull()]
            #print(df_pval.shape)
            df_pval = add_unanno_col(df_pval, all_col='ClusterMembers', in_col='ClusterMembersAnnotated', un_col='ClusterMembersUnannotated')
            df_pval['FDR'] = get_BH_correct_pval(df_pval, pval_col='pval')
            #df_pval_cut = df_pval[df_pval['FDR']<fdr_cutoff]
            df_pval_cut = df_pval.copy(deep=True)
            #df_pval_cut.to_csv(output_sub_dir+'go_enrich_files/Cluster'+str(row[0])+'_GOBP_FDR30.csv')
            df_pval_list.append(df_pval_cut)
            enriched_go = df_pval_cut['GO'].values.tolist()
            num_enriched_go = len(enriched_go)
            enriched_go_str = ';'.join(enriched_go)
            df_clust_new.loc[row[0], 'num_enriched_GO'] = num_enriched_go
            df_clust_new.loc[row[0], 'enriched_GO'] = enriched_go_str
    df_pval_all = pd.concat(df_pval_list)
    df_pval_all.to_csv(output_sub_dir+level+'_clusters_GOBP.csv')
    return df_clust_new, df_pval_all
```

```python
# Write an enrichment function for each condition:
# population = 2k+ non-nan strains 
# success in population = # hits
# sample = # strains annotated to each GO term 
# success in sample: # hits in sample
def go_enrich_condition_hits(df_score, df_go_standard, 
                             condition_list=['FBS', 'Iron', 'Temp37', 'YPD', 'NaCl', 'SDS', 'Sorbitol', 'YNB'],
                            mean_cutoff=2, fdr_cutoff=0.05, output_direc='./'):
    pan_essential_hits = df_score.index
    for cond in condition_list:
        print(cond)
        cond_mean = cond + '_mean'
        cond_fdr = cond + '_fdr'
        df_score_cond_mean = df_score[cond_mean]
        df_score_cond_fdr = df_score[cond_fdr]
        df_score_cond_mean_noNA = df_score_cond_mean[df_score_cond_mean.notnull()]
        df_score_cond_fdr_noNA = df_score_cond_fdr.loc[df_score_cond_mean_noNA.index]
        
        num_population = len(df_score_cond_mean_noNA)
        strains_mean_pass = df_score_cond_mean_noNA[df_score_cond_mean_noNA>=mean_cutoff].index
        strains_fdr_pass = df_score_cond_fdr_noNA[df_score_cond_fdr_noNA<=fdr_cutoff].index
        strains_hit = np.intersect1d(strains_mean_pass, strains_fdr_pass)
        num_hits = len(strains_hit)
        
        pan_essential_hits = np.intersect1d(pan_essential_hits, strains_hit)
        
        print(num_population, num_hits)
        
        enrich_go(df_go_standard, strains_total=df_score_cond_mean_noNA.index, 
                  strains_hit=strains_hit, output_direc=output_direc, condition=cond)
        print()
    
    num_pan_ess_hits = len(pan_essential_hits)
    print('#pan-essential:', num_pan_ess_hits)
    all_genes = df_score.index
    enrich_go(df_go_standard, strains_total=all_genes, 
                  strains_hit=pan_essential_hits, output_direc=output_direc, condition='pan_essential')
    
    
def enrich_go(df_go_standard, strains_total, strains_hit, output_direc, condition):
    num_population = len(strains_total)
    num_hits = len(strains_hit)
    df_go_standard_cond = df_go_standard[['Name', 'Genes']]
    df_go_standard_cond.columns = ['GO_term', 'GO_term_genes']
    df_go_standard_cond['num_strains_total'] = np.nan
    df_go_standard_cond['num_strains_total_GOannotated'] = np.nan
    df_go_standard_cond['num_strain_hits'] = np.nan
    df_go_standard_cond['num_strains_hits_GOannotated'] = np.nan
    df_go_standard_cond['cond_enrich_pval'] = np.nan
    for row in df_go_standard.iterrows():
        go_idx = row[0]
        go_name = row[1]['Name']
        go_genes = row[1]['Genes'].split(';')
        go_in_pop = np.intersect1d(go_genes, strains_total)#df_score_cond_mean_noNA.index)
        go_in_hits = np.intersect1d(go_genes, strains_hit)
        len_go_in_pop = len(go_in_pop)
        len_go_in_hits = len(go_in_hits)
        if len_go_in_hits > 1:
            # We can then compute a probability of drawing X red marbles out of N from 
            # a jar containing n red marbles out of M in the following way:
            df_go_standard_cond.loc[go_idx, 'num_strains_total'] = num_population
            df_go_standard_cond.loc[go_idx, 'num_strains_total_GOannotated'] = len_go_in_pop
            df_go_standard_cond.loc[go_idx, 'num_strain_hits'] = num_hits
            df_go_standard_cond.loc[go_idx, 'num_strains_hits_GOannotated'] = len_go_in_hits

            go_pval = hyper_test(x=len_go_in_hits, M=num_population, n=num_hits, N=len_go_in_pop)
            df_go_standard_cond.loc[go_idx, 'cond_enrich_pval'] = go_pval
    df_go_standard_cond = df_go_standard_cond[df_go_standard_cond['cond_enrich_pval'].notnull()]
    df_go_standard_cond['cond_enrich_fdr'] = get_BH_correct_pval(df_go_standard_cond, pval_col='cond_enrich_pval')
    df_go_standard_cond = df_go_standard_cond.sort_values(by='cond_enrich_fdr', ascending=True)
    if output_direc is not None:
        df_go_standard_cond.to_csv(output_direc+condition+'_go_enrich_results.tsv', sep='\t', index=None)
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

def hierarchical_clustering_within_cluster(df, dist_thre=0.1, metric='precomputed'):
    # Convert correlation matrix to a distance matrix (required for hierarchical clustering)
    correlation_matrix = df.T.corr()
    distance_matrix = 1 - correlation_matrix.abs()

    # Perform hierarchical clustering on the cluster's distance matrix
    cluster_clustering = AgglomerativeClustering(n_clusters=None, distance_threshold=dist_thre, 
                                                 linkage='average', metric=metric)
    if metric == 'precomputed':
        cluster_clustering.fit(distance_matrix)
    else:
        cluster_clustering.fit(df)

    # Sort the genes based on the clustering labels
    df['Cluster_Labels'] = cluster_clustering.labels_
    df = df.sort_values(by='Cluster_Labels')
    
    return df


import seaborn as sns
def cluster_heatmap(df_scores_clust, clust_ID_col='ClusterID', clust_columns=None,
                   subset_clust=None, within_hclust=True, centroid_hclust=False, hclust_dist_thre=0.1,
                    fig_width=30, fig_height=50, cmap_color='OrRd', zscore=True, metric='precomputed',
                   add_clust_gap=False, output_direc=None):
    if clust_columns is None:
        clust_columns = df_scores_clust.columns
        
    #df = df_scores_clust.sort_values(by=clust_ID_col)
    df = df_scores_clust.copy(deep=True)
    
    if subset_clust is not None:
        df = df[df[clust_ID_col].isin(subset_clust)]
        
    # Extract the cluster labels
    clusters = df[clust_ID_col].unique()
    
    # Cluster based on centroid of each cluster
    if centroid_hclust:
        df_clusters = df[[clust_ID_col]].drop_duplicates()
        df_clusters = df_clusters.set_index(clust_ID_col)
        for c in clusters:
            df_c_centroid = df[df[clust_ID_col]==c][clust_columns].mean(axis=0)
            df_clusters.loc[c, clust_columns] = df_c_centroid
        df_clusters = hierarchical_clustering_within_cluster(df_clusters[clust_columns], 
                                                             dist_thre=hclust_dist_thre, metric='euclidean')
        df_list = []
        for c in df_clusters.index:
            df_c = df[df[clust_ID_col]==c]
            df_list.append(df_c)
        df = pd.concat(df_list)
    
    if within_hclust:
        df_list = []
        for c in clusters:
            df_c = hierarchical_clustering_within_cluster(df[df[clust_ID_col]==c][clust_columns], 
                                                          dist_thre=hclust_dist_thre, metric=metric)
            df_list.append(df_c)
        df = pd.concat(df_list)
        
#     if add_clust_gap:
#         df_list = []
#         for c in clusters:
#             df_clust = df[df[clust_ID_col]==c]
#             df_clust.loc[len(df_clust)] = 0 # Add a row of 0's for gap in visualization
#             df_list.append(df_clust)
#         df = pd.concat(df_list)
#         #print(df.shape)
        
    if zscore:
        clust_ID_cols = df[clust_ID_col]
        # Create a StandardScaler object
        scaler = StandardScaler()

        # Fit and transform the data (Z-score normalization)
        df = pd.DataFrame(scaler.fit_transform(df[clust_columns].T), columns=df.index, index=clust_columns)
        df = df.T
        #print(df.mean(axis=1))
        if add_clust_gap:
            df[clust_ID_col] = clust_ID_cols
            df_list = []
            for c in clusters:
                df_clust = df[df[clust_ID_col]==c]
                df_clust.loc[len(df_clust)] = 0 # Add a row of 0's for gap in visualization
                df_list.append(df_clust)
            df = pd.concat(df_list)
            df = df.set_index(clust_ID_col)
        
    # Create a heatmap
    plt.figure(figsize=(fig_width, fig_height))
    if zscore:
        sns.heatmap(df[clust_columns], cmap=cmap_color, 
                    vmin=-3, vmax=3,
                    yticklabels=df.index, xticklabels=clust_columns)
    else:
        sns.heatmap(df[clust_columns], cmap=cmap_color, 
                    yticklabels=df.index, xticklabels=clust_columns)
    # Add cluster labels as y-axis labels
    #plt.yticks(range(len(df)), df.index)

    # Add colorbar
    #plt.colorbar() #label='Attribute Values'
    cluster_str = ''
    for c in clusters:
        cluster_str = cluster_str + str(int(c)) +','
    if cluster_str[-1] == ',':
        cluster_str = cluster_str[:-1]
    plt.title('Heatmap grouped by clusters '+cluster_str)
    plt.xlabel('Effect size per condition')
    plt.ylabel('Gene')

    if output_direc is not None:
        plt.savefig(output_direc, dpi=300)
        
    plt.show()
    
    return df
```

```python
# 09132023: GO enrichment dot plot for interesting sets

def dot_plot_GO_enrich(df_input, go_col='GO_term', fe_col='FoldEnrich', fdr_col='cond_enrich_fdr',
                       size_col='num_strains_hits_GOannotated',
                      fig_width=10, fig_height=10, colorbar_max=20, 
                       just_highlight=True, highlight_col='Highlight', group_name='YPD only',
                      output_direc=None):
    # Sort the DataFrame by p-values
    df = df_input.sort_values(by=fe_col, ascending=True, inplace=False)
    #df[fdr_col] = df[fdr_col] * 100
    
    if just_highlight:
        df = df[df[highlight_col]=='Y']
    
    df.index = list(range(0, len(df)))
    
    plt.figure(figsize=(fig_width, fig_height))
    
    # Scale the size attribute to fit within the desired range
    min_size = 100  # Minimum node size
    max_size = 360  # Maximum node size
    min_size_data = df[size_col].min()
    max_size_data = df[size_col].max()
    df["Scaled_Size"] = np.interp(df[size_col], (min_size_data, max_size_data), (min_size, max_size))

    # Get unique sizes for the legend
    unique_sizes = sorted(df[size_col].unique())
    df_size_map = pd.DataFrame({'size':unique_sizes})
    df_size_map = df_size_map.merge(df[[size_col,'Scaled_Size']],
                      how='left', left_on='size', right_on=size_col)
    df_size_map = df_size_map.drop_duplicates().set_index('size')
    
    # Create a legend with reference nodes for each unique size
    legend_handles = []
    for size in unique_sizes:
        legend_handles.append(plt.scatter([], [], s=df_size_map.loc[size, 'Scaled_Size'], label=f"{size}", color='black')) #, edgecolors="k"

    # Add the legend with reference nodes
    plt.legend(handles=legend_handles, title="# annotated strains", bbox_to_anchor=(2.0, 0.7), loc='upper left')
    
    # Create the dot plot
    plt.scatter(df[fe_col], df.index, c=df[fdr_col], cmap='binary_r', #cmap='viridis', 
                vmin=0, vmax=colorbar_max, s=df['Scaled_Size'])#s=225)#, alpha=0.7)

    # Set y-axis labels as GO term names
    plt.yticks(df.index, df[go_col], fontsize=16)
    plt.xticks(fontsize=16)
    
    # Set labels and title
    plt.xlabel('Fold enrichment', fontsize=16)
    plt.ylabel('GO term', fontsize=16)
    plt.title('Selected enriched GO terms for '+group_name+' hits', fontsize=16, fontweight='bold',loc='left')
    
    plt.xlim([0, df[fe_col].max()+1])
    
    # Move x-axis ticks and labels to the top
    plt.tick_params(axis="y", which="both", right=True, left=False)
    plt.tick_params(labelleft=False, labelright=True)
    
    # Add colorbar to show significance
    #cax = plt.gca().inset_axes([1, 0.1, 0.03, 0.8])  # [left, bottom, width, height]
    cbar = plt.colorbar(pad=0.5, shrink=0.3)#, cax=cax
    cbar.set_label('FDR', fontsize=14)
    plt.clim(vmin=None, vmax=colorbar_max)

    # Show the plot
    plt.grid(True, axis='y', linestyle=(5, (10, 3))) #'long dash with offset'
    plt.tight_layout()
    if output_direc is not None:
        plt.savefig(output_direc, dpi=300)
    plt.show()
```

```python
def get_upper_traingle_pairwise(df):
    # Initialize an empty DataFrame for pairwise data
    pairwise_data = []

    # Loop through the DataFrame to extract upper triangular values
    for i in range(len(df.columns)):
        for j in range(i + 1, len(df.columns)):
            var1 = df.columns[i]
            var2 = df.columns[j]
            correlation = df.loc[var1, var2]
            pairwise_data.append([var1, var2, correlation])

    # Create the pairwise DataFrame
    pairwise_df = pd.DataFrame(pairwise_data, columns=['Replicate 1', 'Replicate 2', 'Correlation'])

    return(pairwise_df)

def get_up_dn_tag_pcc(df_cond_LFC, condition='FBS'):
    df_corr = df_cond_LFC.corr()
    all_reps = df_corr.columns
    up_reps = [c for c in all_reps if '_UP_' in c]
    dn_reps = sorted(list(set(all_reps) - set(up_reps)))
    
    df_corr_up = df_corr.loc[up_reps, up_reps]
    df_corr_dn = df_corr.loc[dn_reps, dn_reps]
    
    df_corr_up_pair = get_upper_traingle_pairwise(df_corr_up)
    df_corr_dn_pair = get_upper_traingle_pairwise(df_corr_dn)
    
    df_corr_up_pair['condition'] = condition
    df_corr_dn_pair['condition'] = condition
    
    #print(df_corr_up_pair['Correlation'].median())
    #print(df_corr_dn_pair['Correlation'].median())
    return df_corr_up_pair, df_corr_dn_pair
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_replicate_pcc(pairwise_df, fig_width=8, fig_height=6,
                      fig_title='Box Plot of Replicate Pairwise Correlations',
                      output_direc=None, swarm_plot=True, swarm_size=3):
    # Create the box plot
    plt.figure(figsize=(fig_width, fig_height))# figsize=(8, 6)
    if swarm_plot:
        sns.swarmplot(x='condition', y='Correlation', data=pairwise_df, size=swarm_size)
        # Add the median value as an annotation
        #median_value = pairwise_df['Correlation'].median()
        #plt.axhline(median_value, color='red', linestyle='--', label=f'Median = {median_value:.2f}')
    else:
        sns.boxplot(x='condition', y='Correlation', data=pairwise_df) 

    # Customize the plot
    plt.title(fig_title)
    #plt.xticks(rotation=45)
    plt.xlabel('Condition')
    plt.ylabel('Correlation')
    #plt.legend()
    plt.ylim([0.0, 1.0])

    # Show the plot
    plt.tight_layout()
    if output_direc is not None:
        plt.savefig(output_direc, dpi=300)
    plt.show()
    
    
def plot_replicate_pcc_updn(pairwise_df, fig_width=8, fig_height=6,
                      fig_title='Box Plot of Replicate Pairwise Correlations',
                      output_direc=None, swarm_plot=True, swarm_size=3,
                           ylim_min=0.0, ylim_max=1.0, font_size=14):
    # Create the box plot
    plt.figure(figsize=(fig_width, fig_height))# figsize=(8, 6)

    if swarm_plot:
        sns.swarmplot(x='condition', y='Correlation', data=pairwise_df, size=swarm_size, hue='tag', dodge=True)
    else:
        sns.stripplot(x='condition', y='Correlation', data=pairwise_df, size=swarm_size, hue='tag', dodge=True)

    # Customize the plot
    if fig_title is not None:
        plt.title(fig_title, fontweight='bold', fontsize=font_size, weight='bold')
    #plt.xticks(rotation=45)
    plt.xlabel('Condition', fontsize=font_size+4, weight='bold')
    plt.ylabel('Correlation', fontsize=font_size+4, weight='bold')
    
    plt.xticks(fontsize=font_size)
    plt.yticks(fontsize=font_size)
    
    #plt.legend()
    plt.ylim([ylim_min, ylim_max])

    # Show the plot
    plt.tight_layout()
    if output_direc is not None:
        plt.savefig(output_direc, dpi=300)
    plt.show()
```
